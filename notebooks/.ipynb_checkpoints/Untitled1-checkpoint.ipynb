```python
from collections.abc import Callable
from functools import partial

import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
#@partial(torch.compile, dynamic=True)
#@partial(torch.jit.script, example_inputs=[(1e-4, 10000, -5.0, 40.0)])
def simulate(h: float, N: int, a: float, f: float) -> torch.Tensor:
    times = torch.arange(0, h * N, h)
    r = torch.ones(N) * h
    theta = torch.ones(N) * h
    h = torch.as_tensor(h)
    for i in range(1, N):
        r[i] = r[i - 1] + h * (a * r[i - 1] - r[i - 1] ** 3) + torch.sqrt(h) * 1e-4 * torch.randn(1)
        theta[i] = theta[i - 1] + h * (2 * np.pi * f) + torch.sqrt(h) * 1e-4 * torch.randn(1)

    return r * torch.exp(1j * theta)
```

```python
N = 50000
h = 1e-4


r = simulate(h, N, torch.tensor(-2), torch.tensor(20))
```

```python
## Load anatomical data
data = np.load("/home/vinicius/Documentos/StuartLandauWBM/interareal/markov2014.npy", allow_pickle=True).item()

# Graph parameters
Nareas = 29  # Number of areas
# FLN matrix
flnMat = data["FLN"]
h = np.squeeze(data["Hierarchy"].T)

flnMat = (1 + 4 * h)[:, np.newaxis] * flnMat
C = (flnMat + flnMat.T) / 2
S = C.sum(1)
```

```python
a_vec = np.linspace(-1, .5, 100)
g_vec = np.linspace(0, 50, 100) / np.linalg.norm(C)

ein = np.zeros((len(a_vec), len(g_vec)))

aprop = np.zeros((len(a_vec), len(g_vec)))


for i, a_0 in enumerate(a_vec):
    print(i)
    for j, g in enumerate(g_vec):

        a = np.random.normal(a_0, 0.2)

        omegas = np.random.normal(2 * np.pi * 40, 2 * np.pi / 10, Nareas)
        
        Axx = Ayy = np.diag(a - .1 * S) + .1 * C
        
        Axy = np.diag(omegas * 2 * np.pi)
        
        Ayx = - Axy
        
        a = np.concatenate((Axx, Axy))
        
        b = np.concatenate((Ayx, Ayy))
        
        A = np.concatenate((a, b), axis=1)
        
        ein[i, j] = np.sign(np.max(np.linalg.eig(A).eigenvalues.real))
        aprop[i, j] = np.sum(a > 0) / Nareas
```

```python
plt.imshow(ein)
plt.colorbar()
```
